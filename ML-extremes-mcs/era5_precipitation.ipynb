```python
import glob
import xarray as xr
```

```python
lsp_list = sorted(glob.glob('/glade/scratch/molina/cesm_mcs/*lsp*nc'))
cps_list = sorted(glob.glob('/glade/scratch/molina/cesm_mcs/*cp*nc'))
```

```python
len(lsp_list) == len(cps_list)
```

```python
for num, i in enumerate(lsp_list):
    tp_ds = xr.open_mfdataset(
        i)['LSP'] + xr.open_mfdataset(cps_list[num])['CP']
    tp_ds.to_dataset(name='TP').to_netcdf(
        i.split('128_142_lsp')[0]+'tp'+i.split('128_142_lsp')[1])
```
